```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
m = 100
X = 6 * np.random.rand(m,1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m,1)
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias= False)
X_poly = poly_features.fit_transform(X)
print(X[0])
```

```python
print(X_poly[0])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
print(lin_reg.intercept_, lin_reg.coef_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data for visualization (replace with your actual data)
X_plot = np.linspace(X.min(), X.max(), 100)  # Adjust range as needed create regression line
X_plot_poly = poly_features.transform(X_plot.reshape(-1, 1))  # Assuming 'poly' is your PolynomialFeatures object
y_plot = lin_reg.predict(X_plot_poly)

# Plot the data points and regression line
plt.scatter(X, y, label='Data Points')
plt.plot(X_plot, y_plot, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression with Polynomial Features')
plt.show()
```
