```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

#resize your graph
plt.figure(figsize=(5,3) ,dpi=300)#dpi use value around 300

#keyword argument notation
#plt.plot(x,y,label ="2x",color = "red",linewidth = 2, marker = ".",markersize = 10,markeredgecolor = "blue",linestyle = "--") #or color = #ababab

#fmt = '[color][marker][line]'
plt.plot(x,y,'r.--',label='2x')

# line number 2
x2 = np.arange(0,4.5,0.5)

#plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'g',label="X^2")

#plot remainder of graph
plt.plot(x2[5:],x2[5:]**2,'r--')

#add a title (specify font paramenters with fontdict)
plt.title("Our first graph",fontdict = {'fontname': 'Comic sans MS','fontsize':20})

#x and y labels
plt.xlabel("X axis")
plt.ylabel("Y axis")

#x and y ticks (scale of the graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.legend()

#save graph will be saved in current directory
plt.savefig('mygraph.png',dpi=300)
plt.show()
```

Bar Charts



```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels , values)


#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch("*")
patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize = (5,3))


plt.show()
```

```python
fifa["Preferred Foot"]
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['Left', 'Right']
colors = ['red','blue']
plt.pie([left,right], labels = labels, colors = colors,autopct= "%.2f %%")
plt.title("Foot Preference of Fifa players")
plt.show()
```

```python
fifa.Weight = fifa.Weight.str.strip("lbs")
```

```python
fifa.dtypes['Weight']
```

```python
fifa['Weight'].fillna(0,inplace = True)
```

```python
fifa['Weight'] = fifa['Weight'].astype(int)
```

```python
fifa.dtypes['Weight']
```

```python
fifa.Weight[0]
```

```python
# fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
light = fifa[fifa.Weight < 125].shape[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight <150)].shape[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight <200)].shape[0]
heavy = fifa[(fifa.Weight >= 200)].shape[0]
weight = [light , light_medium , medium, medium_heavy,heavy]
labels = ["Under 125","125-150","150-175","175-200","Over 200"]
explode = (.4,.1,0,0,.4)
plt.title("Weight Distribution of Fifa players (in lbs)")
plt.pie(weight, labels = labels,autopct = "%.2f %%", pctdistance=0.8, explode= explode)
plt.show()
```

Box and Whiskers 

```python

plt.figure(figsize = (5,8))

barcelona = fifa[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa[fifa.Club == "Real Madrid"]["Overall"]

revs = fifa[fifa.Club == "New England Revolution"]["Overall"]
labels = ["FC Barcelona" ,"Real Madrid","NE Revolution"]
boxes = plt.boxplot([barcelona,madrid,revs], labels=labels,patch_artist = True,medianprops = {'linewidth' :2} )
for box in boxes["boxes"]:
    box.set(color = "red", linewidth =2)

    box.set(facecolor = "#e0e0e0")


plt.title("Professional Soccer Team Comparison")
plt.ylabel("Fifa overall rating")

plt.show()



```
